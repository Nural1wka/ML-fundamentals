```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/mouse_viral_study.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', hue='Virus Present', data=df)

#line
x = np.linspace(0,10,100)

m = -1.1
b = 11

y = m*x + b

plt.plot(x,y, 'black')
```

```python
from sklearn.svm import SVC
```

```python
#help(SVC)
```

```python
y = df['Virus Present']
```

```python
X = df.drop('Virus Present',axis=1)
```

```python
model = SVC(kernel='linear', C=1000)

model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary

plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='linear', C=0.05)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='linear', C=0.005)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=0.1)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma='auto')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma=0.5)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma=0.9)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma=2)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma='scale')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', degree=1)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', degree=2, C=0.05)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', degree=4, C=0.05)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
```

```python
param_grid = {'C': [0.001, 0.01, 0.1, 1],
              'kernel': ['linear', 'rbf']}
```

```python
grid_model = GridSearchCV(svm, param_grid)
```

```python
grid_model.fit(X, y)
```
