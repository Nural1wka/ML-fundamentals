```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
df.head()
```

```python
# Scatter plot of total spend vs sales
# shows a positive correlation
sns.scatterplot(data=df, x='total_spend', y='sales')
```

```python
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
#When we obtain a regression line, we can see that there is a positive correlation
# between total spend and sales. 
#With the knowledge of the regression line, we can predict sales 
# based on total spend.
```

```python
X = df['total_spend']
y = df['sales']
```

```python
# y = mx + b
# y = B1 * X + B0

#help(np.polyfit)
```

```python
np.polyfit(X, y, deg=1)

# B1 = 0.0475
# B0 = 4.243
# y = 0.0475 * X + 4.243
#the coefficients given by np.polyfit are ordered
# from highest degree to lowest degree
# so B1 is the first element and B0 is the second element
```

```python
potential_spend = np.linspace(0, 500, 100)

predicted_sales = 0.04868788 * potential_spend + 4.24302822
```

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')
```

```python
spend = 200

predicted_sales = 0.04868788 * spend + 4.24302822

# If we spend $200, we can expect to sell about 13.98 units
```
