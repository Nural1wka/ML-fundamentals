```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cement_slump.csv')
```

```python
df.head()
```

```python
plt.Figure(figsize=(8, 8), dpi=500)
sns.heatmap(df.corr(), annot=True)
```

```python
df.columns
```

```python
X = df.drop('Compressive Strength (28-day)(Mpa)', axis=1)
```

```python
y = df['Compressive Strength (28-day)(Mpa)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler

#for SVM is is really important to scale (normalize) the features
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVR, LinearSVR
```

```python
#help(SVR)
```

```python
base_model = SVR()
```

```python
base_model.fit(scaled_X_train,y_train)
```

```python
base_pred = base_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print('MAE:',mean_absolute_error(y_test,base_pred))
print('MSE:',mean_squared_error(y_test,base_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,base_pred)))
```

```python
y_test.mean()
```

```python
param_grid = {
    'C': [0.001, 0.01, 0.1, 0.5, 1],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto'],
    'degree': [2, 3, 4],
    'epsilon': [0, 0.01, 0.1, 0.5, 1, 2]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svr = SVR()
```

```python
grid = GridSearchCV(svr, param_grid=param_grid)
```

```python
grid.fit(scaled_X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(scaled_X_test)
```

```python
print('MAE:',mean_absolute_error(y_test,grid_pred))
print('MSE:',mean_squared_error(y_test,grid_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,grid_pred)))
```
